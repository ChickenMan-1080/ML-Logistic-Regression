 #### Target and Encode X

```python
df_X = df.drop(['heart_disease','diabetes','stroke','health_risk_score'],axis=1)
df_X.head()
```

```python
# แปลงค่าใน column ด้วย dictionary
df_X['gender'] = df_X['gender'].map(
    {
    'Male': 1,
    'Female': 0
    }
)

df_X
```

```python
df_y = df[['heart_disease','diabetes','stroke']]
df_y
```

#### Check outlier

```python
#Subplot each Column
fig , axes = plt.subplots(1, 9, figsize=(15, 6))

for ax , col in zip(axes, df.columns):
    df_X.boxplot(column=col, ax=ax)
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

outlier มีลักษณะเป็น Cluster อาจส่งผลต่อ target โดยตรงจึงต้องทำ Bivariate analysis เพื่อดูการเปลี่ยนของข้อมูลเมื่อเกิดค่านอกเกณฑ์ลักษณะกลุ่มก้อน
