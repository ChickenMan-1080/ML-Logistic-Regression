```python
# ดูว่า outlier cluster (BMI 38+) อยู่ใน zone ไหนของ plot จริงๆ
mask = df['bmi'] >= 38

print(df[mask][['bmi', 'blood_pressure', 'gender', 'heart_disease']].describe())

# distribution แยก gender
print(df[mask].groupby(['gender', 'heart_disease']).size())
```

```python
# Plot เพิ่ม marker ทับลงไปเลยเพื่อให้เห็นชัด
outliers = df[df['bmi'] >= 38]

sns.relplot(
    data=outliers,
    x='bmi',
    y='blood_pressure',
    hue='heart_disease',
    col='gender',
    style='gender',
    palette=['g','b'],
    )
```

จากการ plot graph และ สำรวจข้อมูลความสัมพันธ์ของ BMI และ Blood Pressure ที่ส่งผลต่อโรค พบว่า bmi มีผลเล็กน้อยในการเกิดโรคหัวใจบน ความดันเลือดที่ต่ำ อย่างก็ตาม outlier ที่เกิดบน BMI ไม่ได้แสดงผลโดยตรงต่อการเกิดโรคหัวใจอย่างมีนัยสำคัญ

ขณะที่ในเพศหญิง BMI ไม่ได้มีบทบาทสำคัญที่ก่อให้ป่วยเป็นโรคหัวใจ

```python
sns.displot(data=df,x='bmi',y='cholesterol',hue='heart_disease',col='gender',kind='kde')
```

จากการดูเบื้องต้นพบว่า bmi ไม่ได้เป็นตัวแปรสำคัญที่ทำให้เกิดอาการป่วยเป็นโรคหัวใจ
