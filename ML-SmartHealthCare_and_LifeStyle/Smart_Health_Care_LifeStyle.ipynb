```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('assets/smart_healthcare_dataset.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

## **Data Cleaning**

```python
df = df.drop('health_risk_score',axis=1)
df
```

### Check nulls

```python
sns.heatmap(df.isnull())
df.isnull().sum()
```

### Check Dupicate Value

```python
df.duplicated().sum()
```

### Encode

```python
# แปลงค่าใน column ด้วย dictionary
df['gender'] = df['gender'].map(
    {
    'Male': 1,
    'Female': 0
    }
    )

df
```

### Check outlier

```python
#Subplot each Column
fig , axes = plt.subplots(1, 9, figsize=(15, 6))

for ax , col in zip(axes, df.columns):
    df.boxplot(column=col, ax=ax)
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

outlier มีลักษณะเป็น Cluster อาจส่งผลต่อ target โดยตรงจึงต้องทำ Bivariate analysis เพื่อดูการเปลี่ยนของข้อมูลเมื่อเกิดค่านอกเกณฑ์ลักษณะกลุ่มก้อน
